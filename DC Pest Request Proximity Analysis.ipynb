## Collecting Data through `opendata.dc.gov` API

```python
# litter can locations
can_URL = "https://opendata.arcgis.com/datasets/4c9d7e966c0f435485fe0f47144c3258_10.geojson"

can_response = requests.get(can_URL)
#litter_cans = pd.DataFrame(columns = ['object_id', 'rat_cover', 'power_wash', 'geometry'])

litter_can_list = []
for i in range(100):
    obj_id = can_response.json()['features'][i]['properties']['OBJECTID']
    rat = can_response.json()['features'][i]['properties']['RAT_COVER']
    wash = can_response.json()['features'][i]['properties']['POWERWASH']
    geometry = can_response.json()['features'][i]['geometry']
    
    litter_can_list.append({'object_id' : obj_id,
                                      'rat_cover' : rat,
                                      'power_wash' : wash,
                                      'geomtery' : geometry})

litter_cans = pd.DataFrame(litter_can_list)
litter_cans.set_index('object_id', inplace = True)
litter_cans
```

```python
# DC shape
DC_Tile_URL = "https://opendata.arcgis.com/datasets/cea539444b484a5e9034fa3442e2e69a_0.geojson"
DC_response = requests.get(DC_Tile_URL)

# convert to gpd.DataFrame
DC_shape = pd.DataFrame(columns = ['city_name', 'state_name', 'area_km', 'geometry'])

city = DC_response.json()['features'][0]['properties']['CITY_NAME']
state = DC_response.json()['features'][0]['properties']['STATE_NAME']
area = DC_response.json()['features'][0]['properties']['AREAKM']
geom = DC_response.json()['features'][0]['geometry']

DC_shape = DC_shape.append({'city_name' : city,
                           'state_name' : state,
                           'area_km' : area,
                           'geometry': geom}, ignore_index = True)

DC_shape
```

```python
from shapely.geometry import Polygon
polygon_geom = Polygon(DC_shape.geometry[0]['coordinates'][0])
crsa = {'init': 'epsg:4326'}
polygon = gpd.GeoDataFrame({'city_name' : city,
                           'state_name' : state,
                           'area_km' : area},
                           geometry=[polygon_geom],
                           index = [0])
polygon['geometry'][0]
```

```python
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster

m_1 = folium.Map(location=[38.62,-77.0589], tiles='openstreetmap', zoom_start=9)

Choropleth(geo_data=polygon.__geo_interface__, 
           key_on="feature.id", 
           fill_color='teal', 
           legend_name='Major criminal incidents (Jan-Aug 2018)'
          ).add_to(m_1)
m_1
```
